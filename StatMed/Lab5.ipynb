```python
fig = plt.figure(figsize=(9,12))
a = fig.add_subplot(2,1,1)
label.plot.hist(color='lightblue', bins=100)
plt.axvline(label.mean(), color="magenta", linestyle='dashed', linewidth=2)
plt.axvline(label.median(), color="green", linestyle='dashed', linewidth=2)
a.set_title('Histogram')

a = fig.add_subplot(2,1,2)
label.plot(kind='box', vert=False)
a.set_title('Boxplot')
plt.show()
```

```python
numeric_features
```

```python
for col in numeric_features:
    fig = plt.figure(figsize=(10,6))
    ax = fig.gca()
    feature = df[col]
    feature.plot.hist(color='lightblue', bins=100)
    ax.axvline(feature.mean(), color="magenta", linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color="green", linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
categorical_features = ['season','mnth','holiday','weekday','workingday','weathersit','day']
```

```python
for col in categorical_features:
    counts = df[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()
    counts.plot.bar(ax=ax, color='lightblue')
    ax.set_title(col+ ' counts')
    ax.set_xlabel(col)
    ax.set_ylabel("Freq")
plt.show()
```

```python
for col in numeric_features:
    fig =plt.figure(figsize=(9,6))
    ax = fig.gca()
    feature = df[col]
    label = df['rentals']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label, color='lightblue')
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs {} correlation: {:.2f} '.format(col,correlation))
plt.show()
    
```

```python
for col in categorical_features:
    fig = plt.figure(figsize=(10,6))
    ax = fig.gca()
    df.boxplot(column='rentals', by=col, ax=ax)
    ax.set_title('Label by ' + col)
    ax.set_ylabel('Bike Rentals')
plt.show()
```

```python
X = df[numeric_features+['season','mnth','holiday','weekday','workingday','weathersit']].values
y = df['rentals'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(normalize=False)
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
predict = model.predict(X_test)
```

```python
print(f"Predicted values: {np.round(predict)[:10]}")
print(f"True labels: {y_test[:10]}")
```

```python
plt.scatter(y_test, predict, color="lightblue")
plt.xlabel('Actual labels')
plt.ylabel("Predicted labels")
plt.title("Daily Bike Share Predictions")
z = np.polyfit(y_test, predict, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test))
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, predict)
```

```python
mse
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
r2 = r2_score(y_test,predict)
```

```python
r2
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
```

```python
numeric_features = [6,7,8,9]
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_features = [0,1,2,3,4,5]
categorical_transformer = Pipeline(steps=[
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num',numeric_transformer,numeric_features),
        ('cat',categorical_transformer, categorical_features)])

pipline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression(normalize=False))])

```

```python
model = pipline.fit(X_train,y_train)
```

```python
model
```

```python
predict = model.predict(X_test)
mse = mean_squared_error(y_test, predict)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,predict)
print("MSE: {:.2f}\nRMSE:{:.2f}\nR2:{:.2f}".format(mse,rmse,r2))
```

```python
plt.scatter(y_test, predict, color="lightblue")
plt.xlabel('Actual labels')
plt.ylabel("Predicted labels")
plt.title("Daily Bike Share Predictions")
z = np.polyfit(y_test, predict, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test))
plt.show()
```

```python
predict = model.predict(X_test)
mse = mean_squared_error(y_test, predict)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,predict)
print("MSE: {:.2f}\nRMSE:{:.2f}\nR2:{:.2f}".format(mse,rmse,r2))
plt.scatter(y_test, predict, color="lightblue")
plt.xlabel('Actual labels')
plt.ylabel("Predicted labels")
plt.title("Daily Bike Share Predictions")
z = np.polyfit(y_test, predict, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test))
plt.show()
```
